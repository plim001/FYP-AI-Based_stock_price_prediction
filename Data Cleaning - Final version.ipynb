# Data Cleaning

# Li Auto

```python
import pandas as pd
import re
import numpy


LI_Tweets_Clean = pd.read_csv('LI_tweets.csv')
LI_Tweets_Clean = LI_Tweets_Clean.drop_duplicates()
LI_Tweets_Clean['Date2'] = LI_Tweets_Clean['Date'].str[:10]
LI_Tweets_Clean['Time'] = LI_Tweets_Clean['Date'].str[10:19]
LI_Tweets_Clean=LI_Tweets_Clean.drop(LI_Tweets_Clean.columns[0], axis=1)

def cleanTxt(text):
    
    text = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", text)
    text = re.sub(r"#", "", text)                        #Remove '#' symbol
    text = re.sub(r"RT[\s]", "", text)                   #Remove RT  
    text = re.sub(r"-", "", text)                        #Remove '-'
    text = re.sub(r":", "", text)                        #Remove ':'
    text = re.sub(r";", "", text)                        #Remove ';'
    text = re.sub(r",", "", text)                        #Remove ','
    text = re.sub(r"(\n)", "", text)                     #Remove new line in string
    text = re.sub(r"(\W)", " ", text)                    #Remove non-alphanumeric characters
    text = re.sub(' +', ' ', text)    
    return text

#Cleaning the text
LI_Tweets_Clean['Clean_Tweet'] = LI_Tweets_Clean['Tweet'].apply(cleanTxt)

for i in range(len(LI_Tweets_Clean['Clean_Tweet'])):
    LI_Tweets_Clean['Clean_Tweet'][i] = LI_Tweets_Clean['Clean_Tweet'][i].strip()

#Export to csv
LI_Tweets_Clean.to_csv('LI_tweets_Cleaned.csv')

#Show cleaned Text
LI_Tweets_Clean
```
