# Li's Twitter 

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

query = "(@$LI) lang:en until:2022-08-27 since:2020-07-30"                 # Query for LI_auto tweets with date range in twitter
tweets = []
limits = 100000


for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    
    if len(tweets) == limits:
        break
    else:
        tweets.append([tweet.date, tweet.user.username, tweet.content])    # creating columns
        
df = pd.DataFrame(tweets, columns=['Date','User','Tweet'])
df.to_csv('LI_tweets1.csv')                                                # to save to csv
df                                                                         # show what does df contains
```
